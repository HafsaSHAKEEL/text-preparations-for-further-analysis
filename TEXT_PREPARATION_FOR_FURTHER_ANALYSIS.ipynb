```python
# Import Module
from bs4 import BeautifulSoup

# Function to remove tags
def remove_tags(html):

    # parse html content
    soup = BeautifulSoup(html, "html.parser")

    for data in soup(['style', 'script']):
        # Remove tags
        data.decompose()

    # return data by retrieving the tag content
    return ' '.join(soup.stripped_strings)

# Print the extracted data
print(remove_tags(r'C:\Users\Hassaan\Downloads\Documents'))
```

```python
from bs4 import BeautifulSoup
import os

# Define the directory where your documents are stored
doc_directory = r"C:\Users\ammar\Desktop\Slides & Docs\Assignments + Notes\IR\asm1-ir\Documents"

# Create an empty dictionary to store the processed documents
documents = {}

# Loop through each file in the directory
for filename in os.listdir(doc_directory):
    file_path = os.path.join(doc_directory, filename)

    # Read the contents of the file
    with open(file_path, "r",encoding="utf8") as file:
        contents = file.read()

        # Use BeautifulSoup to parse the HTML
        soup = BeautifulSoup(contents, "html.parser")

        # Remove headers and HTML tags from the document
        text = soup.get_text()

        # Store the processed document in the dictionary using the file name as the key
        documents[filename] = text

# You now have a dictionary of documents, where the keys are the file names and the values are the processed text.
```

```python
import os
import re
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt') # download the word tokenizer

def clean_text(file, stop_words):
    # Read the text from the file
    with open(file, "r", encoding="utf8") as f:
        text = f.read()

    # Remove the headers and HTML tags
    text = re.sub("<.*?>", "", text)
    text = re.sub("^.*?\n\n", "", text)

    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove the stop words
    tokens = [token for token in tokens if token.lower() not in stop_words]

    return tokens

# Define the location of the documents
doc_dir = r"C:\Users\ammar\Desktop\Slides & Docs\Assignments + Notes\IR\asm1-ir\Documents"

# Create a hash table to store the cleaned text
doc_dict = {}

# Load the stop words
#with open("stopwords.txt", "r", encoding="utf8") as f:
  #  stop_words = f.read().splitlines()

# Iterate through the files in the directory
for filename in os.listdir(doc_dir):
    file_path = os.path.join(doc_dir, filename)
    if os.path.isfile(file_path):
        # Clean the text and store it in the hash table
        tokens = clean_text(file_path, stop_words)
        doc_dict[filename] = tokens
```
